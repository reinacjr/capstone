```python
# Code to write files to CSV
def write_to_csv(data, filename, header):
    """Writes data to a CSV file with the given header."""
    with open(filename, 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(header)
        writer.writerows(data)
```

```python
## 2. Opportunity 
opp_limit = 10
created_diff = math.floor(created_limit / opp_limit)
prob = [30, 60, 90]

def generate_opportunity_data(leads, names, descs):
    """Generates realistic opportunity data.

    Args:
        num_records (int): The number of opportunity records to generate.
        lead_ids (list, optional): A list of LeadIDs to associate with the opportunities.
                                   If None, random LeadIDs will be generated.

    Returns:
        list: A list of lists, where each inner list represents an opportunity record.
    """

    # add a customer field
    # not all leads becomees opportunities

    stages = ["Discovery", "Connected", "Follow Up"]

    statuses = ["Cancelled", "In Progress", "Won"]

    nameDict = dict(zip(topics, names))
    descDict = dict(zip(topics, descs))
    
    data = []

    count = 1
    for index, row in leads.iterrows():
        # for now per lead can have up to 100 opps 
        created_on = ""
        n = random.randint(1, opp_limit)
        for i in range(n):
            opportunity_id = f"OPP-{count:03}" # identifier for opportunity

            # for now since there isn't a proper table for account yet, will use random int of 4 people
            account_id = f"ACC-{random.randint(1, account_limit):03}" # system generated id, identifier for internal company staff
            lead_id = row.LeadID
            stage = random.choice(stages)
            probability = random.choice(prob) 
        
            estimated_value = random.randint(value_limit[0], value_limit[1])
            
            status = random.choice(statuses) 
            
            topic = row.Topic
            name = nameDict[topic][i]
            description = descDict[topic][i]

            owner_id = row.OwnerID # customer same as in leads

            # datetimes need to be set such that the next record is always larger than the previous record, 
            # if status is won or cancelled, the actual closed date needs to be the same as the modified on 

            if i == 0: # first opportunity 
                created_on = (datetime.now() - timedelta(days=random.randint(0, created_limit))).strftime("%Y-%m-%d")
                
            else:
                created_on = (datetime.strptime(created_on, "%Y-%m-%d") + timedelta(days=random.randint(0, created_diff))).strftime("%Y-%m-%d")

            modified_on = (datetime.strptime(created_on, "%Y-%m-%d") + timedelta(days=random.randint(0, modified_limit))).strftime("%Y-%m-%d")
            expected_close_date = (datetime.strptime(created_on, "%Y-%m-%d") + timedelta(days=random.randint(close_limit[0], close_limit[1]))).strftime("%Y-%m-%d")
            actual_close_date = ""  # Leave blank for In Progress opportunities
            if status == "Won" or status == "Cancelled":
                actual_close_date = modified_on

            data.append([opportunity_id, name, account_id, lead_id, stage, probability, 
                        estimated_value, expected_close_date, actual_close_date, status, 
                        description, owner_id, created_on, modified_on])
            count += 1
    return data


# read leads
leads = pd.read_csv('leads_data.csv')

# filter out leads that have the Converted status
leads = leads[leads['Status'] == "Converted"]

opportunity_data = generate_opportunity_data(leads, opp_names, opp_descs) 

# write data to CSV
write_to_csv(opportunity_data, "opportunities_data.csv", 
             ["OpportunityID", "Name", "AccountID", "LeadID", "Stage", "Probability", "EstimatedValue", 
              "ExpectedCloseDate", "ActualCloseDate", "Status", "Description", "OwnerID", "CreatedOn", "ModifiedOn"])


print(f"Generated {len(opportunity_data)} opportunity records and saved to opportunities_data.csv")
```

```python
# Quote
quote_limit = 10
created_diff = math.floor(quote_limit / opp_limit)
valid_limit = 30
discount_limit = [3, 20]

def generate_quote_data(opportunities):
    """Generates realistic quote data."""

    # statuses = ["Draft", "Sent", "Accepted", "Rejected"]

    data = []

    count = 1
    for index, row in opportunities.iterrows():
        n = random.randint(1, quote_limit)
        created_on = ""
        for i in range (n):
        
            quote_id = f"QUO-{count:03}"
            opportunity_id = row.OpportunityID

            name = row.Name 
            description = row.Description

            account_id = row.AccountID

            quote_number = f"Q{datetime.now().year}-{count:03}" # linked to one AccountID and OpportunityID
            
            total_amount = random.randint(value_limit[0], value_limit[1]) 
            discount_percentage = round(random.uniform(discount_limit[0], discount_limit[1]), 2)  
            discount_amount = round(total_amount * (discount_percentage / 100), 2)
            final_amount = round(total_amount - discount_amount, 2)

            if i == 0: # first quote 
                created_on = (datetime.now() - timedelta(days=random.randint(0, created_limit))).strftime("%Y-%m-%d")
                
            else:
                created_on = (datetime.strptime(modified_on, "%Y-%m-%d") + timedelta(days=random.randint(0, created_diff))).strftime("%Y-%m-%d")

            modified_on = (datetime.strptime(created_on, "%Y-%m-%d") + timedelta(days=random.randint(0, modified_limit))).strftime("%Y-%m-%d")

            valid_from = created_on
            valid_to = (datetime.strptime(created_on, "%Y-%m-%d") + timedelta(days=valid_limit))

            status = ""
            if row.Status == "Cancelled":
                status = "Rejected"
            elif row.Status == "Won":
                if i == n-1:
                    status = "Accepted"
                else:
                    status = "Rejected"
            else: 
                if i == n-1:
                    status = random.choice(["Draft", "Sent"])
                else:
                    status = "Rejected"

            data.append([quote_id, opportunity_id, name, account_id, quote_number, status, 
                        valid_from, valid_to, total_amount, discount_percentage, discount_amount, 
                        final_amount, description, created_on, modified_on])
            count += 1
    return data


opportunities = pd.read_csv('opportunities_data.csv')

# 3. Generate Quote Data, linking to OpportunityIDs and AccountIDs
quote_data = generate_quote_data(opportunities)


write_to_csv(quote_data, "quotes_data.csv", 
             ["QuoteID", "OpportunityID", "Name", "AccountID", "QuoteNumber", "Status", "ValidFrom", "ValidTo", "TotalAmount", "DiscountPercentage", "DiscountAmount", "FinalAmount", "Description", "CreatedOn", "ModifiedOn"])


print(f"Generated {len(quote_data)} quote records.")
```

```python
# Contract 
start_limit = 30
end_limit = [365, 730]
expired_limit = 30

def generate_contract_data(quotes):
    """Generates realistic contract data."""

    statuses = ["Draft", "Pending", "Approved", "Suspended", "Completed", "Cancelled", "Expired"]
    billing_frequencies = ["Monthly", "Quarterly", "Annually", "As Per Contract Milestone"] 
    payment_terms = ["Net 30", "Net 60", "Net 90"]

    data = []
    count = 1
    for index, row in quotes.iterrows():
        contract_id = f"CON-{count:03}"
        quote_id = row.QuoteID
        account_id = row.AccountID
        contract_number = f"C{datetime.strptime(row.ModifiedOn, "%Y-%m-%d").year}-{count:03}"
        status = random.choice(statuses)
        
        total_value = row.FinalAmount
        billing_frequency = random.choice(billing_frequencies)
        payment_term = random.choice(payment_terms)
        name = row.Name + " Contract"
        created_on = row.ModifiedOn # contract created based on when quote was approved
        
        start_date = (datetime.strptime(created_on, "%Y-%m-%d") + timedelta(days=random.randint(0, start_limit))).strftime("%Y-%m-%d") # contract start date within 30 days of quote approval
        end_date = (datetime.strptime(start_date, "%Y-%m-%d") + timedelta(days=random.randint(end_limit[0], end_limit[1]))).strftime("%Y-%m-%d") # 1 to 2 years

        modified_on = (datetime.strptime(created_on, "%Y-%m-%d") + timedelta(days=random.randint(0, modified_limit))).strftime("%Y-%m-%d")

        if status == "Completed": # completed before / on end_date
            modified_on = end_date
        elif status == "Expired": # not completed before end_date
            modified_on = (datetime.strptime(end_date, "%Y-%m-%d") + timedelta(days=random.randint(0, expired_limit))).strftime("%Y-%m-%d")
        elif status == "Suspended": # on hold?
            modified_on = (datetime.strptime(start_date, "%Y-%m-%d") + timedelta(days=random.randint(start_limit, end_limit[1]))).strftime("%Y-%m-%d") # must be suspended between start and end

        data.append([contract_id, quote_id, account_id, contract_number, name, status, start_date, end_date, total_value, billing_frequency, payment_term, created_on, modified_on])

        count += 1
    return data

quotes = pd.read_csv('quotes_data.csv')

quotes = quotes[quotes['Status'] == "Accepted"]

contract_data = generate_contract_data(quotes)

write_to_csv(contract_data, "contracts_data.csv", 
             ["ContractID", "QuoteID", "AccountID", "ContractNumber", "Name", "Status", "StartDate", "EndDate", "TotalValue", "BillingFrequency", "PaymentTerms", "CreatedOn", "ModifiedOn"])

print(f"Generated {len(contract_data)} contract records.")
```

```python
# Milestone
milestone = []
def generate_milestone_data(contracts):
    """Generates realistic contract milestone data."""

    statuses = ["Not Started", "In Progress", "On Hold", "Delayed", "Completed", "Pending Approval", "Under Review", "Cancelled"]

    names = ["Scope Walkthrough", "Reconnaissance and Intelligence Gathering", "Vulnerability Analysis and Exploitation", 
             "Post Exploitation and Impact Assessment", "Reporting and Remediation Recommendation", "Remediation Validation and Retest"]

    data = []

    count = 1
    for index, row in contracts.iterrows():
        due_diff = math.floor((datetime.strptime(row.EndDate, "%Y-%m-%d") - datetime.strptime(row.StartDate, "%Y-%m-%d")).days / len(names)) # to find the due date for each milestone
        due_limit = [-10, +10]
        value_diff = row.TotalValue / len(names)

        current = 0
        if row.Status == "Completed": # if quote status is completed
            current = 6 # set all the milestone statuses as completed
            milestone.insert(index, [current/len(names), row.EndDate])

        else:
            current = random.randint(0, len(names) - 1)

        due_date = ""
        for i in range (len(names)):
            # many milestone to one contract
            # each milestone different status 
            milestone_id = f"MIL-{count:03}"
            contract_id = row.ContractID

            # ! need to define the name here
            name = names[i]
            amount = value_diff
            if i == 0: # if its first milestone
                due_date = (datetime.strptime(row.StartDate, "%Y-%m-%d") + timedelta(days=due_diff)).strftime("%Y-%m-%d")
            else: # if its subsequent milestones
                due_date = (datetime.strptime(due_date, "%Y-%m-%d") + timedelta(days=due_diff)).strftime("%Y-%m-%d")

            created_on = row.StartDate
            modified_on = (datetime.strptime(due_date, "%Y-%m-%d") + timedelta(days=random.randint(due_limit[0], due_limit[1]))).strftime("%Y-%m-%d")

            status = ""
            if i == current: # if milestone is current milestone
                if modified_on > due_date:
                    status = "Delayed"
                elif row.Status == "Suspended":
                    status = "On Hold"
                else:
                    status = "In Progress" 
                milestone.insert(index, [current/len(names), modified_on])
            elif i < current: # if milestone is before current milestone
                status = "Completed"
            elif i > current: # if milestone is after current milestone
                status = "Not Started"
            
            completion_date = ""
            invoice_id = ""
            if status == "Completed":
                completion_date = modified_on
                invoice_id = f"INV-{count:03}"


            data.append([milestone_id, contract_id, name, due_date, amount, 
                        status, completion_date, invoice_id, created_on, modified_on])
            count += 1
    return data

contracts = pd.read_csv('contracts_data.csv')

contracts = contracts[~contracts['Status'].isin(['Draft', 'Pending'])]

milestone_data = generate_milestone_data(contracts)

write_to_csv(milestone_data, "milestones_data.csv", 
             ["MilestoneID", "ContractID", "Name", "DueDate", "Amount", "Status", "CompletionDate", "InvoiceID", "CreatedOn", "ModifiedOn"])

print(f"Generated {len(milestone_data)} milestone records.")
```

```python
# Project 
manager_limit = 3
budget_limit = [8000, 60000]
def generate_project_data(contracts):
    """Generates realistic project data."""

    # statuses = ["Active", "Closed", "Inactive"]
    
    project_managers = [f"USER-{i:03}" for i in range(1, manager_limit)]  # Example Project Managers

    data = []

    count = 1
    for index, row in contracts.iterrows():
        project_id = f"PRJ-{count:03}"
        contract_id = row.ContractID

        name = row.Name + " Project"
        start_date = (datetime.now() + timedelta(days=random.randint(0, 30))).strftime("%Y-%m-%d")
        end_date = (datetime.now() + timedelta(days=random.randint(90, 365))).strftime("%Y-%m-%d")  # At least 3 months duration
        status = ""
        if row.Status == "Approved":
            status = "Active"
        elif row.Status == "Completed":
            status = "Closed"
        else: 
            status = "Inactive"
        project_manager = random.choice(project_managers)
        budget = random.randint(budget_limit[0], budget_limit[1])
        actual_cost = random.randint(0, budget)
        progress = milestone[count-1][0] * 100
        risk = random.choice(["Low", "Medium", "High"])
        created_on = row.StartDate
        modified_on = milestone[count-1][1]

        data.append([project_id, contract_id, name, start_date, end_date, 
                     status, project_manager, budget, actual_cost, progress, risk, 
                     created_on, modified_on])
        count += 1

    return data


contracts = pd.read_csv('contracts_data.csv')

contracts = contracts[~contracts['Status'].isin(['Draft', 'Pending'])]

project_data = generate_project_data(contracts)

write_to_csv(project_data, "projects_data.csv", 
             ["ProjectID", "ContractID", "Name", "StartDate", "EndDate", "Status", "ProjectManager", "Budget", "ActualCost", "Progress", "Risk", "CreatedOn", "ModifiedOn"])

print(f"Generated {len(project_data)} project records.")
```
